# Centrality
Central nodes might impact many other nodes and are useful in fraud to prevent the expansion of future fraudulent activities.

```python
network1 = nx.Graph()
edges = [(1, 2), (2, 3), (1, 3), (3, 5), (2, 5), (5, 6), (5, 7), (3, 4)]
network1.add_edges_from(edges)

plt.figure(figsize=(7,7))
nx.draw_circular(network1, with_labels=True, node_size=1800, font_size=20, alpha=0.6)
```

## Degree Centrality

Node’s degree is used as measure of a node’s degree of connectedness

```python
dc = nx.degree_centrality(network1)

for k, v in dc.items():
    dc[k] = round(v, 2)
    
dc
```

Note: The degree centrality values are normalized by dividing by the maximum possible degree in a simple graph n-1 where n is the number of nodes in G. In the above example, the maximum possible degree is 6.

## Betweeness Centrality
Betweenness centrality of a node 𝑣 is the sum of the fraction of all-pairs shortest paths that pass through 𝑣.

```python
bc = nx.betweenness_centrality(network1, normalized=False)

for k, v in bc.items():
    bc[k] = round(v, 2)
    
bc
```

## Closeness Centrality
Closeness centrality of a node u is the reciprocal of the average shortest path distance to u over all n-1 reachable nodes.

```python
cc = nx.closeness_centrality(network1)

for k, v in cc.items():
    cc[k] = round(v, 2)

cc
```

## Eigenvector Centrality
Eigenvector centrality computes the centrality for a node based on the centrality of its neighbors.

```python
ec = nx.eigenvector_centrality(network1)

for k, v in ec.items():
    ec[k] = round(v, 2)
    
ec
```

## Algorithm to compute clustering coefficient

```python
A = nx.to_numpy_matrix(network6)
```

```python
def clustering_coefficient(A):
    """
    Compute the clustering coefficient of a graph whose adjacency matrix is A
    """
    # Get number of triangles of each node
    trs = np.diag(A**3) / 2
    degrees = np.asarray(A.sum(axis=1)).squeeze()
    max_neighbor_edges = ((degrees * (degrees-1)) / 2) 
    
    # Add a small number to avoid dividing by zero
    max_neighbor_edges += 1e-20
    local_ccs = trs / max_neighbor_edges
    return np.average(local_ccs)
```

```python
round(clustering_coefficient(A), 2)
```

# PageRank
PageRank is an algorithm used by Google Search to rank web pages in their search engine results. 

## Power iteration method without restart probabilities

```python
H = np.array([[0,0.5,1], 
              [1,0,0], 
              [0,0.5,0]])

R = np.random.rand(3,1)

# Normalize the vector such that it is a proper probability distribution
R = R / sum(R)

while True:
    R_new = np.dot(H, R)
    if np.linalg.norm(R_new - R) <= 1e-10:
        break
    R = R_new

print (R.squeeze())
```

## Using dumping factor and restart probabilities

```python
e = np.array([1/3] * 3).reshape(3, 1)
alpha = 0.9

R = np.random.rand(3,1)
# print (R)

# Normalize the vector such that it is a proper probability distribution
R = R / sum(R)

while True:
    R_new = alpha * np.dot(H, R) + (1-alpha) * e
    if np.linalg.norm(R_new - R) <= 1e-10:
        break
    R = R_new

print (R.squeeze())
```

```python
def construct_matrix(sparse_rep, n_dim):
    result = np.zeros((n_dim, n_dim))

    for edge, score in sparse_rep.items():
        result[edge[0]-1, edge[1]-1] = score
        result[edge[1]-1, edge[0]-1] = score
    
    return result
```

```python
betweeness_matrix = construct_matrix(edge_betweenness, 9)
betweeness_matrix*36
```

```python
def findIdx(matrix):
    result = np.where(matrix == np.amax(matrix))
    listOfCordinates = list(zip(result[0], result[1]))
    return result[0][0], result[1][0]
```

```python
G_temp = G2.copy()
nx.draw_circular(G_temp, with_labels=True, node_size=500, alpha=0.6)
plt.show()

for i in range(5):
    edge_betweenness = edge_betweenness_centrality(G_temp)
    bn_matrix = construct_matrix(edge_betweenness, 9)
    x, y = findIdx(bn_matrix)
    print (f'edge to be removed: ({x+1}, {y+1})')
    G_temp.remove_edge(x+1,y+1)
    nx.draw_circular(G_temp, with_labels=True, node_size=500, alpha=0.6)
    plt.show()
```
